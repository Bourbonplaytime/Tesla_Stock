```python
import pandas as pd
import sqlite3
from matplotlib import pyplot as plt
import numpy as np
import datetime
```

```python
data = pd.read_csv('TSLA.csv', parse_dates=True)

data
```

```python
plt.subplots(figsize = [9, 6])

data['High'].plot(label='High')

data['Low'].plot(label='Low')

data['Close'].plot(label='Close')

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Tesla Price Over Time (06-2010 to 02-2020)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data['Year'] = data['Date'].dt.year
data[['Year', 'Volume', 'Close']].groupby('Year').sum().plot()
```
