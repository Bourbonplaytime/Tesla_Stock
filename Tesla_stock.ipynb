```python
import pandas as pd
import sqlite3
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('TSLA.csv')

data
```

```python
graph = plt.figure(figsize=[12, 9])

graph.legend()

index = data['Date']

ax = graph.add_subplot(1, 1, 1)

ax.plot(data['High'])

ay = graph.add_subplot(1, 1, 1)

ay.plot(data['Low'])

az = graph.add_subplot(1, 1, 1)

az.plot(data['Close'])
```

```python
plt.subplots(figsize = [9, 6])

dates = data['Date']

index = np.arange(11)

width = 0.5

plt.plot(dates, data['High'], label='High')

plt.plot(dates, data['Low'], label='Low')

plt.plot(dates, data['Close'], label='Close')

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Tesla Price Over Time (06-2010 to 02-2020)')
plt.xticks([])
plt.legend()

plt.tight_layout()
plt.show()
```
